    <h1 align="left"><font color='green'>2. Types of Regression</font></h1>
    <a class='anchor' id='types'></a> 
    <a href="#top" style="color:#36FF00;">[↑] Back to Top</a>
</div>

Regression analysis is a powerful tool for understanding relationships between variables and making predictions. Different types of regression are used depending on the nature of the data and the relationship between variables.

<h3 align="left"> <font color='red'>1. Linear Regression</font></h3>    


**Linear Regression** models the relationship between a dependent variable and one or more independent variables using a straight line. It is the simplest form of regression analysis.

- **Formula**: $y = \beta_0 + \beta_1 x + \epsilon$
- **Use Case**: Predicting continuous outcomes, like house prices or stock values.

<h3 align="left"> <font color='red'>2. Multiple Linear Regression</font></h3>     

**Multiple Linear Regression** extends linear regression by modeling the relationship between a dependent variable and multiple independent variables.

- **Formula**: $y = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n + \epsilon$
- **Use Case**: Examining the effect of several factors on a dependent variable.

<h3 align="left"> <font color='red'> 3. Polynomial Regression</font></h3>

**Polynomial Regression** models non-linear relationships by introducing polynomial terms. It can capture the curvature in the data.

- **Formula**: $y = \beta_0 + \beta_1 x + \beta_2 x^2 + \ldots + \beta_n x^n + \epsilon$
- **Use Case**: Situations where the relationship between variables is non-linear.

<h3 align="left"> <font color='red'>4. Ridge Regression</font></h3>

**Ridge Regression** is a type of linear regression that includes a regularization term to prevent overfitting. It adds a penalty equivalent to the square of the magnitude of coefficients.

- **Objective Function**:  
  $$\text{Minimize} \ \sum (y_i - \beta_0 - \beta_1 x_i)^2 + \lambda \sum \beta_j^2$$
- **Use Case**: When multicollinearity is present in the data.

<h3 align="left"> <font color='red'>5. Lasso Regression</font></h3> 

**Lasso Regression** (Least Absolute Shrinkage and Selection Operator) is another form of regularization that adds a penalty equivalent to the absolute value of the magnitude of coefficients.

- **Objective Function**:  
  $$\text{Minimize} \ \sum (y_i - \beta_0 - \beta_1 x_i)^2 + \lambda \sum |\beta_j|$$
- **Use Case**: Feature selection and reducing model complexity.

<h3 align="left"> <font color='red'>6. Logistic Regression</font></h3> 

**Logistic Regression** is used for binary classification problems. It models the probability that a given input belongs to a particular category.

- **Formula**:  
  $$P(y=1) = \frac{1}{1 + e^{-(\beta_0 + \beta_1 x)}}$$
- **Use Case**: Classifying outcomes like spam detection, and churn prediction.

<h3 align="left"> <font color='red'>7. Elastic Net Regression</font></h3>  

**Elastic Net Regression** combines both Ridge and Lasso regularizations. It can be useful when there are multiple correlated variables.

- **Objective Function**:  
  $$\text{Minimize} \ \sum (y_i - \beta_0 - \beta_1 x_i)^2 + \lambda_1 \sum \beta_j^2 + \lambda_2 \sum |\beta_j|$$
- **Use Case**: When both feature selection and multicollinearity are concerns.

<h3 align="left"> <font color='red'>8. Support Vector Regression (SVR)</font></h3> 

**Support Vector Regression** is based on the principles of support vector machines. It tries to fit the best line within a predefined threshold.

- **Use Case**: Non-linear relationships with high-dimensional data.

<h3 align="left"> <font color='red'>9. Decision Tree Regression</font></h3> 

**Decision Tree Regression** uses a tree structure to model decisions and possible consequences. It splits the data into branches based on feature values.

- **Use Case**: Handling non-linear relationships and capturing complex data patterns.

* <h3 align="left"> <font color='red'>Definition & Working principle</font></h3>    

* * Linear regression is a **supervised learining** algorithm used when target / dependent variable  **continues** real number. It establishes relationship between dependent variable $y$ and one or more independent variable $x$ using best fit line.   It work on the principle of ordinary least square $(OLS)$ / Mean square errror $(MSE)$. In statistics ols is method to estimated unkown parameter of linear regression function, it's goal is to minimize sum of square difference between observed dependent variable in the given data set and those predicted by linear regression fuction. 

* <h3 align="left"> <font color='red'>Hypothesis representation</font></h3>    

* * We will use $\mathbf{x_i}$ to denote the independent variable and $\mathbf{y_i}$ to denote dependent variable. A pair of $\mathbf{(x_i,y_i)}$ is called training example. The subscripe $\mathbf{i}$ in the notation is simply index into the training set. We have $\mathbf{m}$ training example then $\mathbf{i = 1,2,3,...m}$.

* * The goal of supervised learning is to learn a *hypothesis function $\mathbf{h}$*, for a given training set that can used to estimate $\mathbf{y}$ based on $\mathbf{x}$. So hypothesis fuction represented as 

$$\mathbf{ h_\theta(x_{i}) = \theta_0 + \theta_1x_i }$$   
* * $\mathbf{\theta_0,\theta_1}$ are parameter of hypothesis.This is equation for **Simple / Univariate Linear regression**. 

* * For **Multiple Linear regression** more than one independent variable exit then we will use $\mathbf{x_{ij}}$ to denote indepedent variable and $\mathbf{y_{i}}$ to denote dependent variable. We have $\mathbf{n}$ independent variable then $\mathbf{j=1,2,3 ..... n}$. The hypothesis function represented as

$$\mathbf{h_\theta(x_{i}) = \theta_0 + \theta_1x_{i1} + \theta_2 x_{i2} + ..... \theta_j x_{ij} ...... \theta_n  x_{mn} }$$
* * $\mathbf{\theta_0,\theta_1,....\theta_j....\theta_n }$ are parameter of hypothesis,
$\mathbf{m}$ Number of training exaples,
$\mathbf{n}$ Number of independent variable,
$\mathbf{x_{ij}}$ is $\mathbf{i^{th}}$ training exaple of $\mathbf{j^{th}}$ feature.

    <h1 align="left"><font color='green'>4. Logistic Regression</font></h1>
    <a class='anchor' id='logistic'></a> 
    <a href="#top" style="color:#36FF00;">[↑] Back to Top</a>
</div>
Logistic regression estimates the probability of an event occurring, such as voting or not voting, based on a given dataset of independent variables. This type of statistical model, also known as the **logit model**, is often used for classification and predictive analytics. Since the outcome is a probability, the dependent variable is bounded between 0 and 1.

<h3 align="left"> <font color='red'>Key Concepts</font></h3>    

- **Logit Transformation**: In logistic regression, a logit transformation is applied to the odds, which is the probability of success divided by the probability of failure. This is known as the log odds or the natural logarithm of odds.

- **Logistic Function**: The logistic function is represented by the following formulas:

  1. **Logistic Function**:
     \[
     $\text{Logit}(p) = \frac{1}{1 + \exp(-p)}$
     \]

  2. **Logit Transformation**:
     \[
     $\ln\left(\frac{p}{1-p}\right) = \beta_0 + \beta_1 X_1 + \ldots + \beta_k X_k$
     \]

<h3 align="left"> <font color='red'>Explanation</font></h3>    


- **Dependent Variable**: In this logistic regression equation, $\text{logit}(p)$ is the dependent or response variable, and $X$ represents the independent variables.

- **Coefficients**: The $\beta$ parameters (or coefficients) in this model are commonly estimated via maximum likelihood estimation (MLE). This method tests different values of $\beta$ through multiple iterations to optimize the best fit of log odds.

- **Log Likelihood Function**: All these iterations produce the log likelihood function, and logistic regression seeks to maximize this function to find the best parameter estimates.

<h3 align="left"> <font color='red'>Predictive Analysis</font></h3>   

- **Probability Calculation**: Once the optimal coefficients are found, the conditional probabilities for each observation can be calculated, logged, and summed to yield a predicted probability.

- **Binary Classification**: For binary classification, a probability less than 0.5 predicts 0 (failure), while a probability greater than 0.5 predicts 1 (success).

<h3 align="left"> <font color='red'>Model Evaluation</font></h3>   

- **Goodness of Fit**: After computing the model, it’s best practice to evaluate how well the model predicts the dependent variable, which is called the goodness of fit.

- **Hosmer–Lemeshow Test**: This is a popular method to assess model fit, ensuring that the model accurately represents the observed data.

    <h1 align="left"><font color='green'>5. Overfitting and Underrfitting</font></h1>
    <a class='anchor' id='over'></a> 
    <a href="#top" style="color:#36FF00;">[↑] Back to Top</a>
</div>

Accuracy and error are the two other important metrics. The error is the difference between the actual value and the predicted value estimated by the model. Accuracy is the fraction of predictions our model got right. 

For a model to be ideal, it’s expected to have low variance, low bias and low error. To achieve this, we need to partition the dataset into train and test datasets. The model will then learn patterns from the training dataset and the performance will be evaluated on the test dataset. To reduce the error while the model is learning, we come up with an error function which will be reviewed in the following section. If the model memorizes/mimics the training data fed to it, rather than finding patterns, it will give false predictions on unseen data. The curve derived from the trained model would then pass through all the data points and the accuracy on the test dataset is low. This is called __overfitting__ and is caused by high variance. 

On the flip side, if the model performs well on the test data but with low accuracy on the training data, then this leads to __underfitting__.

    <a href="#top" style="color:#36FF00;">[↑] Back to Top</a>

<h3 align="left"> <font color='red'>Introduction</font></h3>    

Machine learning is an iterative process.

You will face choices about what predictive variables to use, what types of models to use, what arguments to supply to those models, etc. So far, you have made these choices in a data-driven way by measuring model quality with a validation (or holdout) set.

But there are some drawbacks to this approach. To see this, imagine you have a dataset with 5000 rows. You will typically keep about 20% of the data as a validation dataset, or 1000 rows. But this leaves some random chance in determining model scores. That is, a model might do well on one set of 1000 rows, even if it would be inaccurate on a different 1000 rows.

At an extreme, you could imagine having only 1 row of data in the validation set. If you compare alternative models, which one makes the best predictions on a single data point will be mostly a matter of luck!

In general, the larger the validation set, the less randomness (aka "noise") there is in our measure of model quality, and the more reliable it will be. Unfortunately, we can only get a large validation set by removing rows from our training data, and smaller training datasets mean worse models!

<h3 align="left"> <font color='red'>What is cross-validation?</font></h3>    

In cross-validation, we run our modeling process on different subsets of the data to get multiple measures of model quality.

For example, we could begin by dividing the data into 5 pieces, each 20% of the full dataset. In this case, we say that we have broken the data into 5 "folds".



<center>
<img src="src/9k60cVA.png" >
</center>

Then, we run one experiment for each fold:

* In Experiment 1, we use the first fold as a validation (or holdout) set and everything else as training data. This gives us a measure of model quality based on a 20% holdout set.
* In Experiment 2, we hold out data from the second fold (and use everything except the second fold for training the model). The holdout set is then used to get a second estimate of model quality.
* We repeat this process, using every fold once as the holdout set. Putting this together, 100% of the data is used as holdout at some point, and we end up with a measure of model quality that is based on all of the rows in the dataset (even if we don't use all rows simultaneously).

<div class="alert alert-block alert-info">

<b>Correlation</b> is a statistical measure that expresses the strength of the relationship between two variables.

The two main types of correlation: positive and negative.

Positive correlation occurs when two variables move in the same direction; as one increases, so do the other.

Negative correlation occurs when two variables move in opposite directions; as one increases, the other decreases.

Correlation can be used to test hypotheses about cause-and-effect relationships between variables. Correlation is often used in the real world to predict trends.

```python
#.corr() use to find correlation with data
data_corr= df1.drop(columns=['Seasons','Holiday']).corr()       
plt.figure(figsize=(12,12))
#cmap use for colour palette , linewidths : set the width of the lines that divide the cells, 
#annot :helps in annotating the heatmap with values if set to True, otherwise values are not provided , linecolor : helps in setting the color of each line that divides heatmap cells
sns.heatmap(data_corr, cmap='coolwarm', linewidths=0.1, annot=True, linecolor='white')
```

```python
data_corr
```

```python
# Find pairs of highly correlated features
high_corr_pairs = []

threshold = 0.8

for i in range(len(data_corr.columns)):
    for j in range(i):
        if abs(data_corr.iloc[i, j]) > threshold:
            colname1 = data_corr.columns[i]
            colname2 = data_corr.columns[j]
            high_corr_pairs.append((colname1, colname2, data_corr.iloc[i, j]))

# Display highly correlated feature pairs
for pair in high_corr_pairs:
    print(f"{pair[0]} and {pair[1]}: Correlation = {pair[2]:.2f}")
```


<div class="alert alert-block alert-info">

When considering which features to remove based on correlation, you typically look for:

1. High Correlation Between Features: If two features are highly correlated with each other, it may indicate multicollinearity, which can inflate the variance of the coefficient estimates and make the model unstable. A common threshold for high correlation is above 0.8 or below -0.8.

2. Low Correlation with Target: Features that have very low correlation with the target variable might not contribute much to the predictive power of the model. However, correlation alone should not be the only metric to decide this, as some features may have non-linear relationships with the target.

```python
def add_lag_features(df, rolling = 1):
    df = df.copy()
    
    # add the history of demand in the last hour
    df['demand_lag_1hr'] = df['Rented Bike Count'].rolling(rolling, closed="left").mean()
    df['temperature_lag_1hr'] = df['Temperature(°C)'].rolling(rolling, closed="left").mean()
    df['rainfall_lag_1hr'] = df['Rainfall(mm)'].rolling(rolling, closed="left").mean()
    df['snowfall_lag_1hr'] = df['Snowfall (cm)'].rolling(rolling, closed="left").mean()
    
    
    df = df.dropna()
    
    return df
```

### Introduction

Scaling is a crucial step in the data preprocessing pipeline, especially when working with machine learning algorithms that are sensitive to the magnitude of feature values, such as gradient descent-based algorithms (e.g., linear regression, logistic regression) and distance-based algorithms (e.g., k-nearest neighbors, k-means clustering). By scaling your features, you ensure that each feature contributes equally to the distance computations or gradient updates.

### Common Scaling Techniques

1. **Min-Max Scaling (Normalization)**
   - Scales the data to a fixed range, typically [0, 1].
   - Formula: $X_{\text{scaled}} = \frac{X - X_{\text{min}}}{X_{\text{max}} - X_{\text{min}}}$

2. **Standardization (Z-score Scaling)**
   - Scales the data such that the mean of the transformed data is 0 and the standard deviation is 1.
   - Formula: $X_{\text{scaled}} = \frac{X - \mu}{\sigma}$
   - Useful for algorithms that assume normally distributed data.

3. **Robust Scaling**
   - Uses the median and interquartile range for scaling, making it robust to outliers.
   - Formula: $X_{\text{scaled}} = \frac{X - \text{median}}{\text{IQR}}$

4. **MaxAbs Scaling**
   - Scales each feature by its maximum absolute value, preserving the sparsity of the data.

<div class="alert alert-block alert-info">

We will predict value for target variable by using our model parameter for test data set. Then compare the predicted value with actual valu in test set. We compute **Mean Square Error** using formula 
$$\mathbf{ J(\theta) = \frac{1}{m} \sum_{i=1}^{m}(\hat{y}_i - y_i)^2}$$

$\mathbf{R^2}$ is statistical measure of how close data are to the fitted regression line. $\mathbf{R^2}$ is always between 0 to 100%. 0% indicated that model explains none of the variability of the response data around it's mean. 100% indicated that model explains all the variablity of the response data around the mean.

$$\mathbf{R^2 = 1 - \frac{SSE}{SST}}$$
**SSE = Sum of Square Error**  
**SST = Sum of Square Total**  
$$\mathbf{SSE = \sum_{i=1}^{m}(\hat{y}_i - y_i)^2}$$
$$\mathbf{SST = \sum_{i=1}^{m}(y_i - \bar{y}_i)^2}$$
Here $\mathbf{\hat{y}}$ is predicted value and $\mathbf{\bar{y}}$ is mean value of $\mathbf{y}$.

</div>



```python
# sklearn regression module
y_pred = lin_reg.predict(X_test)

#Evaluvation: MSE
J_mse = mean_squared_error(y_pred, y_test)

# R_square
R_square = lin_reg.score(X_test,y_test)
print('The Mean Square Error(MSE) or J(theta) is: ',J_mse)
print('R square obtain for scikit learn library is :',R_square)
```

<h3 align="left"> <font color='red'> Model Validation </font></h3>    


<div class="alert alert-block alert-info">

In order to validated model we need to check few assumption of linear regression model. The common assumption for *Linear Regression* model are following
1. Linear Relationship: In linear regression the relationship between the dependent and independent variable to be *linear*. This can be checked by scatter ploting Actual value Vs Predicted value
2. The residual error plot should be *normally* distributed.
3. The *mean* of *residual error* should be 0 or close to 0 as much as possible
4. The linear regression require all variables to be multivariate normal. This assumption can best checked with Q-Q plot.
5. Linear regession assumes that there is little or no *Multicollinearity in the data. Multicollinearity occurs when the independent variables are too highly correlated with each other. The variance inflation factor *VIF* identifies correlation between independent variables and strength of that correlation. $\mathbf{VIF = \frac {1}{1-R^2}}$, If VIF >1 & VIF <5 moderate correlation, VIF < 5 critical level of multicollinearity.
6. Homoscedasticity: The data are homoscedastic meaning the residuals are equal across the regression line. We can look at residual Vs fitted value scatter plot. If heteroscedastic plot would exhibit a funnel shape pattern.

</div>

```python
# Check for Multivariate Normality

# Quantile-Quantile plot 
f,ax = plt.subplots(1,2,figsize=(14,6))
import scipy as sp
_,(_,_,r)= sp.stats.probplot((y_test - y_pred),fit=True,plot=ax[0])
ax[0].set_title('Check for Multivariate Normality: \nQ-Q Plot')

#Check for Homoscedasticity
sns.scatterplot(y = (y_test - y_pred), x= y_pred, ax = ax[1],color='r') 
ax[1].set_title('Check for Homoscedasticity: \nResidual Vs Predicted');
```
